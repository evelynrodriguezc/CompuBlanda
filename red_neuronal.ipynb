## Una capa de neuronas

Las redes neuronales suelen tener capas que constan de más de una neurona. Las capas no son nada
más que grupos de neuronas. Cada neurona de una capa recibe exactamente la misma entrada: la entrada
dada a la capa (que pueden ser los datos de entrenamiento o la salida de la capa anterior),
pero contiene su propio conjunto de pesos y su propio sesgo (bias), produciendo su propia salida única. 

La capa de salida es un conjunto de cada una de estas salidas, una por cada neurona. Supongamos que tenemos un escenario con 3 neuronas en una capa y 4 entradas:

neuronas-03.PNG

```python
inputs = [1, 2, 3, 2.5]

weights1 = [0.2, 0.8, -0.5, 1]
weights2 = [0.5, -0.91, 0.26, -0.5]
weights3 = [-0.26, -0.27, 0.17, 0.87]

bias1 = 2
bias2 = 3
bias3 = 0.5

outputs = [
  # Neuron 1:
  inputs[0]*weights1[0] +
  inputs[1]*weights1[1] +
  inputs[2]*weights1[2] +
  inputs[3]*weights1[3] + bias1,

  # Neuron 2:
  inputs[0]*weights2[0] +
  inputs[1]*weights2[1] +
  inputs[2]*weights2[2] +
  inputs[3]*weights2[3] + bias2,

  # Neuron 3:
  inputs[0]*weights3[0] +
  inputs[1]*weights3[1] +
  inputs[2]*weights3[2] +
  inputs[3]*weights3[3] + bias3]

print (outputs)
```

```python
inputs = [1, 2, 3, 2.5]

weights = [
           [0.2, 0.8, -0.5, 1],
           [0.5, -0.91, 0.26, -0.5],
           [-0.26, -0.27, 0.17, 0.87]
          ]

biases = [2, 3, 0.5]

# Salida de la capa actual
layer_outputs = []

# Para cada neurona
for neuron_weights, neuron_bias in zip(weights, biases):

  # Salida cero de una neurona dada
  neuron_output = 0

  # Por cada entrada y peso a la neurona
  for n_input, weight in zip(inputs, neuron_weights):

    # Multiplique esta entrada por el peso asociado
    # y agrégelo a la variable de salida de la neurona
    neuron_output += n_input*weight

  # Se agrega el bias
  neuron_output += neuron_bias

  # Coloca el resultado de la neurona en la lista de salida de la capa.
  layer_outputs.append(neuron_output)
  
print (layer_outputs)
```
